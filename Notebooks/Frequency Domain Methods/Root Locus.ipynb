```python
# Imports 
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
import control.matlab as cmatlab
#import slycot
from scipy import signal

# Plotting settings (to format plots nicely)
plt.rcParams['figure.figsize'] = [8, 8]
plt.rcParams.update({'font.size': 18})

# Define s as being the transfer function Laplace variable
s = cmatlab.tf('s')
```

```python
G = 1 / (s**2 + 1.5*s + 1)
cmatlab.pzmap(G)

plt.plot()
```

```python
for bn in range(1, 11):
    b = bn/10
    G = 1 / (s**2 + b*s + 1)
    cmatlab.pzmap(G, label="b=" + str(b))
plt.plot()
```

```python
# G = (s**2 + s + 1) / (s**3 + 4*s**2 + 1*s + 1)
G = s / (s**3 + 4*s**2 + 3)
cmatlab.rlocus(G)
plt.plot()
```
