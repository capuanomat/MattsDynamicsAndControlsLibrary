```python
# Imports 
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
import control.matlab as cmatlab
#import slycot
from scipy import signal

# Plotting settings (to format plots nicely)
plt.rcParams['figure.figsize'] = [8, 8]
plt.rcParams.update({'font.size': 18})

# Define s as being the transfer function Laplace variable
s = cmatlab.tf('s')
```

```python
sys = cmatlab.tf([1, 3, 5, 7], [2, 4, 6, 8, 2])
print("Open-Loop Poles and Zeros")
cmatlab.pzmap(sys)
plt.plot()
```

```python
print("Closed-Loop Poles and Zeros")
cmatlab.pzmap(1+sys)
plt.plot()
```

```python
real, imag, freq = cmatlab.nyquist(sys, label="Open-loop")
real, imag, freq = cmatlab.nyquist(sys + 1, label="Closed-loop")
```
